# Data Pre - Processing

```python
#dropping the ID coloumn since it is unnecessary
loan_data.drop(['ID'], axis =1, inplace = True)
```

```python
# "Experience' coloumn has negative values, we will replace those with the median values because the distribution is skewed
loan_data['Experience'].mask(loan_data['Experience'] < 0, loan_data.Experience.median(), inplace=True)

#Checking the, modified values now 
loan_data.describe().transpose()
```

```python
loan_data['ZIP Code'].value_counts()   #checking zip code distribution

# Distributing zip codes on the basis of probabilities so that we can extract some information out of it



zip_inv = loan_data.groupby(by='ZIP Code')['Personal Loan'].agg([('total','count'),('positive','sum')]).sort_values(['total'],ascending=False)

zip_inv['ratio']=zip_inv['positive']/zip_inv['total'] 

# for doing this we will categorise zip codes in two bins on the basis of percentage of "1" in the target varible

zip_inv['bins']  = pd.qcut(zip_inv.ratio,q=2)  

# Now we have two intervals of ratio-  1st one is (0 - 0.04) and other one is (0.04 - 0.667)

zip_inv['bins'].value_counts() 
zip_inv.groupby(by='bins').sum()
zip_inv.bins.unique()
map1={'(0.04, 0.667]':0, '(-0.001, 0.04]':1}
zip_inv['map_col']=zip_inv['bins'].astype('str').map(map1)

map2= zip_inv['map_col']  
# Now we will append this extra information extracted from "ZIP Code" to a new coloumn named as "map_zip"

loan_data['map_zip'] =loan_data['ZIP Code'].map(map2)
```

```python
# Now that we have extracted information from ZIP Code and mapped it into the dataset int the form of a coloumn "map_zip", we don't require 'ZIP Code' in the data set. Therefore we will drop it.
loan_data.drop(['ZIP Code'], axis =1, inplace = True)  
```

# Splitting the data into training and test set in the ratio of 70:30 respectively

```python
X = loan_data.drop('Personal Loan',axis=1)    
Y = loan_data['Personal Loan']   

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=1)

x_train.head()
```

```python
#Checking the split of data
print("{0:0.2f}% data is in training set".format((len(x_train)/len(loan_data.index)) * 100))
print("{0:0.2f}% data is in test set".format((len(x_test)/len(loan_data.index)) * 100))
```

# Training a Logistic Regression Model

```python
from sklearn import metrics

from sklearn.linear_model import LogisticRegression

# Fit the model on train
model = LogisticRegression(solver="liblinear")
model.fit(x_train, y_train)
#predict on test
y_predict = model.predict(x_test)


coef_df = pd.DataFrame(model.coef_)
coef_df['intercept'] = model.intercept_
print(coef_df)
```

### Performance of model with testing data

```python
model_score = model.score(x_test, y_test)
print(model_score)
```

### Confusion Matrix

```python
cm=metrics.confusion_matrix(y_test, y_predict, labels=[1, 0])

df_cm = pd.DataFrame(cm, index = [i for i in ["1","0"]],
                  columns = [i for i in ["Predict 1","Predict 0"]])
plt.figure(figsize = (7,5))
sns.heatmap(df_cm, annot=True,fmt='d')
```

# Training a KNN model

```python
#importing libraries
from sklearn.neighbors import KNeighborsClassifier
```

```python
NNH = KNeighborsClassifier(n_neighbors= 5 , weights = 'distance' )
NNH.fit(x_train, y_train)
```

### performance of model with testing data

```python
# For every test data point, predict it's label based on 5 nearest neighbours in this model. The majority class will 
# be assigned to the test data point

predicted_labels = NNH.predict(x_test)
NNH.score(x_test, y_test)
```

### Confusion Matrix

```python
cm=metrics.confusion_matrix(y_test, predicted_labels, labels=[1, 0])

df_cm = pd.DataFrame(cm, index = [i for i in ["1","0"]],
                  columns = [i for i in ["Predict 1","Predict 0"]])
plt.figure(figsize = (7,5))
sns.heatmap(df_cm, annot=True,fmt='d')
```

# Training a  Naive Bayes Model

```python
#importing necessary libraries
from sklearn.naive_bayes import GaussianNB # using Gaussian algorithm from Naive Bayes

# creatw the model
loan_model = GaussianNB()

loan_model.fit(x_train, y_train.ravel())
```

### Performance of our model with testing data


```python
test_predict = loan_model.predict(x_test)

from sklearn import metrics

print("Model Accuracy: {0:.4f}".format(metrics.accuracy_score(y_test, test_predict)))
print()
```

### Confusion Matrix

```python
print("Confusion Matrix")
cm=metrics.confusion_matrix(y_test, test_predict, labels=[1, 0])

df_cm = pd.DataFrame(cm, index = [i for i in ["1","0"]],
                  columns = [i for i in ["Predict 1","Predict 0"]])
plt.figure(figsize = (7,5))
sns.heatmap(df_cm, annot=True, fmt ='d')
```
